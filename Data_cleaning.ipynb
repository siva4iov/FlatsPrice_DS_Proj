#Rename columns
First of all, let's rename the columns so that the names would be in дatin

```python
new_names = [
    "Balcony", "Furniture", "Address", "Living_space", "Num_of_rooms", 
 "Renovation", "Internet_tv", "Price", "Appliances", 'Total_area', 
 'Kitchen_area', "Bathroom", "Floor", "Utilities", 'Room_type'
 ]
mapper = {key:value for key, value in zip(df.columns.values, new_names)}

df = df.rename(columns=mapper)
df
```

#Count None entries
Let's count none values, and all none rows

```python
def count_none():
  for col in df.columns:
    nasum = df[col].isna().sum()
    print(f'{col} - {nasum} none entries')
count_none()
none_rows = np.all(df.isna().values, axis=1)
num_none_rows = none_rows.sum()
print(f'Full none rows - {num_none_rows}')
```

**As we can see, we have 44 completely none rows.**  
If we get rid of them, then each row will have a price, a total_area etc.

# Let's drop them

```python
df = df[~none_rows]
count_none()
```

Much better!
