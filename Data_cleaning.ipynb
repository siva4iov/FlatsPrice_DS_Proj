### Im going to start with `Appliances`

My idea is to convert values to arrays, and then create column for each unique element in this arrays and for each row put `1` where this value in array, else 0





```python
def convert_appliances(df: pd.DataFrame) -> pd.DataFrame:
  series = df.loc[:, "Appliances"]
  data = []
  unique_values = set()
  new_cols_data = []

  for row in series.values:
    row_c = [value.strip() for value in (row.split(",") if isinstance(row, str) else [])]
    for element in row_c:
      if not element in unique_values:
        unique_values.add(element)
    data.append(row_c)
  
  unique_values = list(unique_values)

  for row in data:
    new_cols_data.append([1 if value in row else 0 for value in unique_values])
  
  col_names = [f"Appliances_{i+1}" for i in range(len(unique_values))]
  print("\n".join(map(str,zip(unique_values, col_names))))
  
  # df.loc[:, "Appliances"] = data
  df.drop(columns=["Appliances"], axis=1, inplace=True)

  additional_cols = pd.DataFrame(data=new_cols_data, columns=col_names, dtype=np.int8)

  return pd.concat([df, additional_cols], axis=1)

df = convert_appliances(df)
```

When we have dealt with the most difficult case. We can do with others the same thing(`Renovation` is exception, we will just ordinal ecnode it).

Also, there are in `Internet_tv` column two values "*телевидение*" и 
"*телевизор*" that mean the same thing. Whe should handle this/

```python
def convert_lists(df: pd.DataFrame) -> pd.DataFrame:
  cols = ["Balcony", "Furniture", "Internet_tv", 
"Bathroom", "Utilities", 'Room_type'
 ]
  new_cols = []
  for col in cols:
    series = df.loc[:, col]
    if col == "Internet_tv":
      series = series.str.replace("телевизор", "телевидение")
    data = []
    unique_values = set()
    new_cols_data = []

    for row in series.values:
      row_c = [value.strip() for value in (row.split(",") if isinstance(row, str) else [])]
      for element in row_c:
        if not element in unique_values:
          unique_values.add(element)
      data.append(row_c)
    
    unique_values = list(unique_values)

    for row in data:
      new_cols_data.append([1 if value in row else 0 for value in unique_values])
    
    col_names = [f"{col}_{i+1}" for i in range(len(unique_values))]
    new_cols.append(pd.DataFrame(data=new_cols_data, columns=col_names, dtype=np.int8))
    print("\n".join(map(str,zip(unique_values, col_names))))

    df.drop(columns=[col], axis=1, inplace=True)

  return pd.concat([df]+new_cols, axis=1)

df = convert_lists(df)
```
